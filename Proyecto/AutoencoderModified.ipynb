```python
import numpy as np
import pandas as pd

def split_into_quadrants(lst, original_grid_size=97, quadrant_size=33):
    # Reshape the list into a 97x97 grid, padding if necessary
    padded_length = quadrant_size * 3  # 99
    padded_grid = np.full((padded_length, padded_length), lst[-1])
    original_grid = np.reshape(lst, (original_grid_size, original_grid_size))
    padded_grid[:original_grid_size, :original_grid_size] = original_grid

    # Extract 9 quadrants of 33x33
    quadrants = []
    for i in range(0, padded_length, quadrant_size):
        for j in range(0, padded_length, quadrant_size):
            quadrant = padded_grid[i:i+quadrant_size, j:j+quadrant_size].flatten()
            quadrants.append(quadrant)

    return quadrants

# New DataFrame data
new_rows = []

for _, row in df.iterrows():
    split_row = {col: split_into_quadrants(row[col]) for col in df.columns}
    for i in range(9):  # 9 quadrants
        new_rows.append([split_row[col][i] for col in df.columns])

# Create the new DataFrame
new_df = pd.DataFrame(new_rows, columns=df.columns)

# Example usage with your DataFrame
df = new_df
```

```python
maps = []

for index, row in df.iterrows():
    #Create a list to merge the arrays in the row and store the values
    merged_list = []
    for column in df.columns:
        merged_list.extend(row[column])
    #maps.append(row['json.layerHeight'])
    maps.append(merged_list)
```

```python
import keras_tuner as kt
from keras.layers import Input, Dense,Flatten, Reshape
from keras.models import Model
from keras.callbacks import EarlyStopping, ModelCheckpoint

#CALLBACKS
cp = ModelCheckpoint('modelWorking_12_20_2023.keras', monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
es = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=5, restore_best_weights=True)

#NEURONE NUMBER
#hp_units = kt.HyperParameter('units', min_value=32, max_value=512, step=32)
size1= 1000*6
size2 = 500*6
size3 = 100*6

sequence_length = len(maps[0]) #How many values are in each array
num_classes = len(maps[0][0]) #How many classes are in each array

# Assuming dim1, dim2, and dim3 represent your input dimensions
input_img = Input(shape=(sequence_length,num_classes))

# Encoder with dense layers
x = Flatten()(input_img)
x = Dense(units=size1, activation='relu')(x)
x = Dense(units=size2, activation='relu')(x)
encoded = Dense(units=size3, activation='relu')(x)

# Decoder with dense layers
latent_shape = Input(shape=(size3))
x = Dense(units=size2, activation='relu')(latent_shape)
x = Dense(units=size1, activation='relu')(x)
x = Dense(units=sequence_length*num_classes, activation='relu')(x)
x = Reshape(target_shape=(sequence_length,num_classes))(x)
decoded = Dense(num_classes, activation='softmax')(x)

encoder = Model(input_img, encoded)
decoder = Model(latent_shape, decoded)

# Autoencoder
autoencoder = Model(input_img, decoder(encoded))
autoencoder.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])

autoencoder.summary()
```

```python
autoencoder.fit(np.array(maps), np.array(maps), epochs=100, batch_size=32, shuffle=True, validation_split=0.2, callbacks=[cp, es])
```

```python
import numpy as np
import pandas as pd

def reassemble_quadrants(quadrants, original_grid_size=97, quadrant_size=33):
    padded_length = quadrant_size * 3  # 99
    padded_grid = np.zeros((padded_length, padded_length))

    quadrant_index = 0
    for i in range(0, padded_length, quadrant_size):
        for j in range(0, padded_length, quadrant_size):
            quadrant = quadrants[quadrant_index].reshape(quadrant_size, quadrant_size)
            padded_grid[i:i+quadrant_size, j:j+quadrant_size] = quadrant
            quadrant_index += 1

    return padded_grid[:original_grid_size, :original_grid_size].flatten()

# Assuming new_df is your modified DataFrame

original_rows = []
for i in range(0, len(decoded_df), 9):  # Process every 9 rows
    reassembled_row = {col: reassemble_quadrants([decoded_df.iloc[j][col] for j in range(i, i + 9)]) 
                       for col in decoded_df.columns}
    original_rows.append(reassembled_row)

# Create the original DataFrame
original_df = pd.DataFrame(original_rows)

```
