```python
#Open json file from dir and convert to numpy array
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load original data
#Search for all json files in the directory
path_to_dir = r'C:\Users\claud\Desktop\ExtractedMaps'
#path_to_dir = r'D:\DSI_PCG\ConvertedMaps'

#Create an array to store the names of the json files in the directory
json_files = [pos_json for pos_json in os.listdir(path_to_dir) if pos_json.endswith('.json')]
#print(json_files)

#Create a dataframe to store the data from the json files
original = pd.DataFrame()

#Loop through each name in the json_files array and load the data into the dataframe
for index, js in enumerate(json_files):
    #with open(os.path.join(path_to_dir, js)) as json_file:
    with open(os.path.join(path_to_dir, js), 'r', encoding='utf-8') as json_file:
        json_text = json.load(json_file)
        #user concat method to append the dataframes together
        values = pd.json_normalize(json_text)
        original = pd.concat([original, values], ignore_index=True)
original.head()
```

```python
#Remove last 6 columns of original dataframe
original = original.iloc[:, :-6]
original.head()
```

Load Decoded Data

```python

#Search for all json files in the directory
path_to_dir = r'D:\DSI_PCG\OutputMaps'
#Create an array to store the names of the json files in the directory
json_files = [pos_json for pos_json in os.listdir(path_to_dir) if pos_json.endswith('.json')]
#Create a dataframe to store the data from the json files
newValues = pd.DataFrame()

#Loop through each name in the json_files array and load the data into the dataframe
for index, js in enumerate(json_files):
    #with open(os.path.join(path_to_dir, js)) as json_file:
    with open(os.path.join(path_to_dir, js), 'r', encoding='utf-8') as json_file:
        json_text = json.load(json_file)
        #user concat method to append the dataframes together
        values = pd.json_normalize(json_text)
        newValues = pd.concat([newValues, values], ignore_index=True)
newValues.head()
```

```python
#Transpose the dataframe
newValues = newValues.T
#Change column names to 	json.flags	json.groundTexture	json.groundVariation	json.cliffVariation	json.cliffTexture	json.layerHeight
newValues.columns = ['json.flags', 'json.groundTexture', 'json.groundVariation', 'json.cliffVariation', 'json.cliffTexture', 'json.layerHeight']
```

```python
newValues.head()
```

Change old Data by new Data

```python
#Create a list to store the original column names
originalColumns = original.columns.tolist()

#Create a list to store the newValues column names
newValuesColumns = newValues.columns.tolist()

#Merge the two lists together
allColumns = originalColumns + newValuesColumns


#Convert the original dataframe to a numpy array
original = original.to_numpy()

#Convert the newValues dataframe to a numpy array
newValues = newValues.to_numpy()

#Create a new array with all the column names
allColumns = np.concatenate((originalColumns, newValuesColumns), axis=0)

merged = np.concatenate((original, newValues), axis=1)

#Convert the merged array to a dataframe
merged = pd.DataFrame(merged)

#Change the column names of the dataframe
merged.columns = allColumns

merged
```
